```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('C:/Users/Daniel/Pictures/Ejemplo.jpg',0)

s = []
for i in range(100):
    for j in range(100):
        s.append(img[i,j])
s = np.array(s)
fig, ax = plt.subplots()
ax.plot(s)

for i in range(10):
    s = np.abs(s - np.mean(s))
fig, ax = plt.subplots()
ax.plot(s)

s = s > 20
fig, ax = plt.subplots()
ax.plot(s)

ax.grid()


plt.show()
```
